## Loading and preprocessing the data

```python
# # Load data
(data_x, _), _ = mnist.load_data()
data_x = np.reshape(np.asarray(data_x), [60000, 28, 28, 1]).astype(float)

# Parameters
no, rows, cols, _ = data_x.shape

# Introduce missing data
data_x, data_m, mask = data_preprocessor(data_x, 0.1, 28, 28)
```

```python
conv_gain_parameters = {'batch_size':128,
                        'hint_rate':0.1,
                        'learning_rate':0.0001,
                        'alpha':0.05,
                        'iterations':4}

imputed_data = Conv_GAIN(data_m, conv_gain_parameters)
```
